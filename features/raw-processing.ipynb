```python
import os 
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/home/juan/Documents/Master_Project/data/process/dataset_all_12-03.csv')
df  = data.iloc[:,1:]
df.head()
```

```python

columns_to_sum = {
    '': ['VDWaals', 'ELEC', 'HBONds', 'ASP', 'USER'],
    '_1': ['VDWaals_1', 'ELEC_1', 'HBONds_1', 'ASP_1', 'USER_1'],
    '_4': ['VDWaals_4', 'ELEC_4', 'HBONds_4', 'ASP_4', 'USER_4'],
    '_5': ['VDWaals_5', 'ELEC_5', 'HBONds_5', 'ASP_5', 'USER_5'],
    '_6': ['VDWaals_6', 'ELEC_6', 'HBONds_6', 'ASP_6', 'USER_6'],
    '_7': ['VDWaals_7', 'ELEC_7', 'HBONds_7', 'ASP_7', 'USER_7']
}

# Iterate over the suffixes
for suffix, cols in columns_to_sum.items():
    # Sum the corresponding columns and replace the original columns with the summed values
    df[f'MM{suffix}'] = df[cols].sum(axis=1)
    # Drop the original columns
    df.drop(columns=cols, inplace=True)

print(df)
```
